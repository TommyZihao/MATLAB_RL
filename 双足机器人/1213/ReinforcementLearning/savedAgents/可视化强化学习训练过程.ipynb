## 导入工具包

```python
import scipy.io as scio

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
```

## 导入训练日志数据csv文件

```python
csv_filename = 'trainingResults_3D_12_01_2022_1643.csv'
df = pd.read_csv(csv_filename, names=['EpisodeIndex','EpisodeReward','EpisodeSteps','AverageReward','TotalAgentSteps','AverageSteps','EpisodeQ0'])
```

```python
df
```

## 获取训练日志数据

```python
EpisodeIndex = df['EpisodeIndex']
EpisodeReward = df['EpisodeReward']
AverageReward = df['AverageReward']
EpisodeQ0 = df['EpisodeQ0']

EpisodeSteps = df['EpisodeSteps']
TotalAgentSteps = df['TotalAgentSteps']
AverageSteps = df['AverageSteps']
```

## 可视化训练过程中的奖励

```python
plt.figure(figsize=(14,8))
plt.plot(EpisodeIndex, EpisodeReward, label='Episode Reward')
plt.plot(EpisodeIndex, AverageReward, label='Average Reward')
plt.plot(EpisodeIndex, EpisodeQ0, label='Episode Q0')

plt.axhline(y=0, color='k', linewidth=0.8) # 绘制 X 轴

plt.legend(fontsize='large')
plt.title('Episode Traning', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.plot(TotalAgentSteps, EpisodeReward, label='Episode Reward')
plt.plot(TotalAgentSteps, AverageReward, label='Average Reward')
plt.plot(TotalAgentSteps, EpisodeQ0, label='Episode Q0')

plt.axhline(y=0, color='k', linewidth=0.8) # 绘制 X 轴

plt.legend(fontsize='large')
plt.title('Step Traning', fontsize=20)
plt.show()
```

## 可视化智能体存活时间

```python
plt.figure(figsize=(14,8))
plt.plot(EpisodeIndex, EpisodeSteps, label='Episode Steps')
plt.plot(EpisodeIndex, AverageSteps, label='Average Steps')
plt.legend(fontsize='large')

plt.title('Survive Steps', fontsize=20)

# 图表左下角设为(0,0)点
# plt.xlim(left=0)
# plt.ylim(bottom=0)
plt.show()
```
